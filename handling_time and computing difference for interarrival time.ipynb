#### How do you take the difference of one column in a data frame
The following example should help 

```python
arrivals = [{'FltNo': 'UA435', 'ArrTime': 150},
         {'FltNo': 'UA478', 'ArrTime': 245},
         {'FltNo': 'UA234', 'ArrTime': 213},
         {'FltNo': 'UA761', 'ArrTime': 116},
         {'FltNo': 'UA981', 'ArrTime': 445},
         {'FltNo': 'UA123', 'ArrTime': 110},         
         {'FltNo': 'UA001', 'ArrTime': 106}]
mydf = pd.DataFrame(arrivals)
mydf
```

```python
mydfsorted = mydf.sort_values(by="ArrTime",ascending=1)
#mydfsorted = mydfsorted.reset_index(drop=True)
mydfsorted
```

```python
mydfsorted['InterArrivalTime'] = mydfsorted['ArrTime'].diff()
mydfsorted
```
